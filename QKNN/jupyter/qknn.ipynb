```python
import sys
sys.path.append("/home/ptsapce/AISeQ-Lab/QKNN")

import numpy as np
from base.qknn import HybridQKNN
```

```python
Dtrain = np.array([[1,0], [1,1], [0,1], [1,0]])

Ytrain = np.array([0, 0, 1, 1])



Dtest = np.array([[0.2, 0.1], [0.9, 0.8]])
```

```python
hqknn = HybridQKNN(k=2, niter=3, shots=1024)
hqknn.fit(Dtrain, Ytrain)
```

```python
predictions = hqknn.predict(Dtest)
print("Predictions:", predictions)
```

```python
import matplotlib.pyplot as plt

distances_matrix = np.zeros((len(Dtest), len(Dtrain)))
for i, vi in enumerate(Dtest):
    for j, uj in enumerate(Dtrain):
        distances_matrix[i, j] = hqknn.quantum_distance(uj, vi)

plt.imshow(distances_matrix, cmap='viridis', interpolation='nearest')
plt.colorbar(label="Quantum distance")
plt.xlabel("Training index")
plt.ylabel("Test index")
plt.title("Hybrid QkNN Quantum Distances")
plt.show()
```
